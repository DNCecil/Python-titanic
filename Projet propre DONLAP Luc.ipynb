### ***Traitons les variables abérrantes des trains et des tests***

***___Commençons avec les train___***

```python
'''détection des valeurs abérrantes en utilisant la méthode interquartile'''
def out_iqr(df , column):
    global borne_sup, borne_inf
    q25, q75 = np.quantile(df[column], 0.25), np.quantile(df[column], 0.75)
    
    # calcul de l'EI
    ei = q75 - q25
    
    # limite d'outlier
    cut_off = ei * 3
    
    # on calcule les bornes inf et sup pour un outlier
    borne_inf, borne_sup = q25 - cut_off, q75 + cut_off
    print("L'ecart interquartile est ", ei)
    print('La borne inférieur est ', borne_inf)
    print('La borne supérieur est ', borne_sup)
    
    # on calcule le nombre d'outliers
    df1 = df[df[column] > borne_sup]
    df2 = df[df[column] < borne_inf]
    return print("Nombre total d'outliers: ", df1.shape[0]+ df2.shape[0])
```

```python
out_iqr(data_train, 'SibSp')
```

```python
'''détection des valeurs abérrantes en utilisant la méthode écart type'''
def out_std(df, column):
    global borne_inf, borne_sup
    
    # calcul de la moyenne et de l'écart type
    data_mean, data_std = df[column].mean(), df[column].std()
    
    # calcul de la valeur limite
    cut_off = data_std * 3
    
    # calcul des bornes superieures et inférieures 
    borne_inf, borne_sup = data_mean - cut_off, data_mean + cut_off
    
    print('La borne inf est ', borne_inf)
    print('La borne sup est ', borne_sup)
    # Calcul du nombre de valeurs au dessus et en dessous des bornes
    df1 = df[df[column] > borne_sup]
    df2 = df[df[column] < borne_inf]
    return print("Nombre total d'outliers:", df1.shape[0]+ df2.shape[0])
```

```python
out_std(data_train, 'SibSp')
```

```python
'''détection des valeurs abérrantes en utilisant le z score'''
def out_zscore(data):
    global outliers, zscore
    outliers = []
    zscore = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score= (i - mean)/std 
        zscore.append(z_score)
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return print("Total number of outliers are",len(outliers))
```

```python
out_zscore(data_train.SibSp)
```

***Ayant comparer les différentes méthodes de détection de valeurs aberrantes sur la variable 'SibSp', j'ai choisi de continuer avec la méthode de l'écart-type parce qu'elle présente un même résultats que la méthode du z-score et me permet d'avoir une certaine fiabilité sur la détection des valeurs manquantes.***

***Je vais commencer à traiter d'abord les valeurs manquantes qui se trouvent dans embarked en les imputant avec le mode calculé de la colonne Embarked***

```python
# Calculez le mode de la colonne
mode_Embarked = data_train['Embarked'].mode().iloc[0]

# Remplacez les valeurs manquantes par le mode
data_train['Embarked'].fillna(mode_Embarked, inplace=True)

# Affichez le DataFrame mis à jour
print(data_train['Embarked'])
```

***Poursuivons avec les valeurs manquantes qui se trouvent dans Age en les imputant en fonction de la Pclass à l'aide des boxplot ci-dessus.***

Pour le faire, nous allons utiliser l'age moyen en fonction des classes pour imputer les valeurs manquantes se trouvant dans Age.

```python
def impute_age(cols) : 
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
data_train['Age'] = data_train[['Age','Pclass']].apply(impute_age,axis=1)
```

***___Standardisation des variables indépendantes (X)___***

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

***___Construction et entrainement du modèle de classification___***

Modele Random Forest Classifier

```python
#Entrainons le modèle
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Prédictions sur le X_train
y_pred = model.predict(X_train)
score = accuracy_score(y_pred, y_train)

print(f"Mon modèle a une précision de {np.round(score*100, 2)}%")
```

```python
# Prédictions sur le X_test
y_pred = model.predict(X_test)
score = accuracy_score(y_pred, y_test)

print(f"Mon modèle a une précision de {np.round(score*100, 2)}%")
```
